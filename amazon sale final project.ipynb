# Changing 'order date' & 'ship date' in date format

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
data.head()
```

# Adding new column 'month' & 'year'

```python
data['Month'] = data['Order Date'].dt.month
data['Year']  = data['Order Date'].dt.year
data.head()
```

# Draw correlation matrix

```python
plt.figure(figsize=(12,6))
sns.heatmap(data.corr(),annot = True)
plt.show()
```

# Sales chaneel

```python
sns.countplot(data=data, x= 'Sales Channel')
plt.title('Sales channel')
plt.show()
```

# Region oders

```python
sns.countplot(data = data, y = 'Region')
plt.title('REGION ORDERS')
plt.show()
```

# Item type sales

```python
sns.countplot(data = data, y = 'Item Type')
plt.title('item type sales')
plt.show()
```

# Month wise oders

```python
sns.countplot(data = data, x = 'Month')
plt.title('MONTH WISE ORDERS')
plt.show()
```

# Year wise oders

```python
sns.countplot(data = data, x = 'Year')
plt.title('YEAR WISE ORDERS')
plt.show()
```

# Total revenue per year

```python
data.groupby(data['Order Date'].dt.year)['Total Revenue'].sum().plot(kind='bar')
plt.title('Total revenue year-wise')
plt.show()
```

# Total revenue per month

```python
data.groupby(data['Order Date'].dt.month)['Total Revenue'].sum().plot(kind='bar')
plt.title('Total revenue month-wise')
plt.show()
```

# Total revenue Year & month-wise

```python
plt.figure(figsize=(16,8))
data.groupby([data['Order Date'].dt.year, data['Month']])['Total Revenue'].sum().unstack().plot(kind='bar')
plt.title('Total revenue Year & month-wise')
plt.show()
```
